# **Seaborn**


```python
import seaborn as sns
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("community_health_metric.csv")
```

```python
df 
```

```python
df.info()
```

```python
df.describe()
```

## Scatter plot (dağılım grafiği)

```python
sns.scatterplot(data=df,x="kilo",y="boy")
```

```python
plt.figure(figsize=(8,6),dpi=200)
sns.scatterplot(df,x="kilo",y="boy",hue="cinsiyet") #hue 3. parametreye göre ayrıştırma yapıyor.
```

```python
plt.figure(figsize=(8,6),dpi=200)
sns.scatterplot(df,x="kilo",y="boy",hue="cinsiyet",size="etnik") # etnik derecelerine göre boyutlandırdık daireleri.
```

```python
plt.figure(figsize=(8,6),dpi=200)
sns.scatterplot(df,x="kilo",y="boy",hue="cinsiyet",alpha= 0.5,s=90,palette="viridis") #alpha soluklaştırma işlemi yapıyor.
```

```python
plt.figure(figsize=(8,6),dpi=200)
sns.scatterplot(df,x="kilo",y="boy",hue="cinsiyet",alpha= 0.5,s=90,palette="viridis",linewidth=0.1) #linewidth ile aralarındaki beyaz çizgileri yok edebiliriz.
```

```python
plt.figure(figsize=(8,6),dpi=200)
sns.scatterplot(df,x="kilo",y="boy",hue="cinsiyet",alpha= 0.8,s=90,palette="viridis",style="cinsiyet") #stili cinsiyete göre yaptığımız için şekilleri değişti.
```
