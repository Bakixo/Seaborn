## Bar plot

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("community_health_metric.csv")
```

```python
sns.barplot(df,x="dr",y="kilo") # burada örneğin 17 defa doktora gidenlerin ortalaması yaklaşık 90'dır diyebilmemizi sağlıyor
                                # siyah çizgiler ise standart sapmayı gösteriyor.
```

```python
sns.barplot(df,x="etnik",y="boy")
plt.ylim(150,190)
```

```python
sns.barplot(df,x="etnik",y="boy",ci="sd") # standart sapmayı böylede ekstra alabiliriz.
plt.ylim(150,190)
```

```python
sns.barplot(df,x="etnik",y="boy",ci="sd",estimator=np.min) # en küçüğüne göre sırala dedik.
```

```python
sns.barplot(df,x="etnik",y="boy",ci="sd",hue="cinsiyet")
plt.ylim(130,200)
```

```python
sns.barplot(df,x="etnik",y="boy",ci="sd",hue="cinsiyet")
plt.ylim(130,200)
plt.legend(bbox_to_anchor=(1,1))
```

## Histogram

```python
sns.displot(df,x="dr")
```

```python
sns.histplot(df,x="dr")
```

```python
sns.histplot(df,x="dr",bins=10) # bins=10, 10 tane çubuk ver.
```

```python
sns.histplot(df,x="dr",bins=10,kde=True) #kde ağırlaştırılmış bir eğri yerleştirir.
```

```python
sns.set(style="whitegrid")
sns.histplot(df,x="dr",bins=10,kde=True)
```

```python
sns.histplot(df,x="dr",bins=10,kde=True,edgecolor="red",
             lw=2.5)
```

```python
sns.histplot(df,x="dr",kde=True,edgecolor="red",
             lw=2.5,ls="--",color="black")
```

## Rugplot

```python
sns.rugplot(df,x="yas",height=0.5) # rugplot sık olan yerlere çizgi çekmeye yarıyor. height ile yüksenliğini ayarlayabiliriz.
```

```python
sns.kdeplot(df,x="yas") # kdeplot daha çok dağılımın ne tarafa daha çok yığıldığına bakmak için kullanılıyor.
```

```python
# az önceki tabloları birleştirelim.

sns.displot(df,x="yas",kde=True,rug="True",bins=15,
            )
```

## KDE

```python
sns.kdeplot(df,x="yas",bw_adjust=1) # kendisi hali hazırda bw default değeri 1 olduğundan grafik değişmedi. 
```

```python
sns.kdeplot(df,x="yas",bw_adjust=8) # çok düzgün hale geldi.
```

```python
sns.kdeplot(df,x="yas",bw_adjust=0.1) # çok fazla hassas oldu şimdi de.
```

```python
sns.kdeplot(df,x="yas",bw_adjust=0.7,shade=True) # shade içini doldurmamızı sağlar.
```

```python
sns.kdeplot(df,x="yas",bw_adjust=0.7,shade=True,color="orange")
```

```python
sns.kdeplot(df,x="yas",y="kilo") # halkaların içerisine girdikçe yoğunluk artıyor demek.
```
