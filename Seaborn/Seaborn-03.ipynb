## Boxplot


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("community_health_metric.csv")
```

```python
sns.boxplot(df,x="etnik",y="kilo")
"""
karelerin içerisindeki bölünen çizgi %50 sini gösterir. yani medyan değeri.
alt tarafı alt %25, üst tarafı da üst %25 si 
üst çizgi ve alt çizgiler de 1.5 katı olacak şekilde çizilir.

"""
```

```python
sns.boxplot(df,y="etnik",x="kilo", orient="h")
```

```python
sns.boxplot(data=df,x="etnik",y="kilo",hue="cinsiyet",width=0.6)
plt.legend(bbox_to_anchor=(1,0.5))
```

## Violinplot

```python
sns.violinplot(df,y="dr",x="etnik")
```

```python
sns.violinplot(df,y="dr",x="etnik",hue="cinsiyet")
```

```python
sns.violinplot(df,y="dr",x="etnik",hue="cinsiyet",split=True)
```

```python
sns.violinplot(df,y="dr",x="etnik",hue="cinsiyet",split=True,
                inner="quartile")
plt.legend(bbox_to_anchor=(1,0.5)) #burada inner çizgileri kadın ve erkeklerin ayrı medyanlarını gösteriyor.
```
